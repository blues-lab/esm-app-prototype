```python
import pandas as pd
import json
from datetime import datetime
from dateutil import parser as dateparser
import csv
pd.options.display.max_colwidth=1000
```

```python
since = datetime(2019, 8, 21)#date since when the data should be kept
log_keys = ['Type','File','Function','Message','Time']
```

```python
json_data = json.load(open('InvitationCode.json'))
invitation_codes = dict()
for data in json_data:
    if dateparser.parse(data['received']).date()>= since.date():
        invitation_codes[data['uid']] = json.loads(data['value'])['InvitationCode']
    
invitation_codes
```

```python
log_data = json.load(open('Log.json'))
log_data=[r for r in log_data if dateparser.parse(r['received']).date()>= since.date()]
logs = []
for r in log_data:
    try:
        identifiers = {'uuid':r['uid'],'received':r['received'],'client':r['client'],
                       'invitation_code':invitation_codes[r['uid']]}
        for line in r['value'].split('\n'):
            if(len(line)>0):
                if(isinstance(line,str)):
                    js = json.loads(line)
                else:
                    js = line
                js.update(identifiers)
                logs.append(js)   
    except Exception as e:
        print(e)
        print(r['value'])
        break
```

```python
log_df = pd.DataFrame.from_dict(logs)
log_df.set_index(['invitation_code','uuid'], inplace=True)
log_df.head()
```
