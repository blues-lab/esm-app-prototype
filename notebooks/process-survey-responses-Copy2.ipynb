```python
import pandas as pd
import json
from datetime import datetime
from dateutil import parser as dateparser

pd.options.display.max_colwidth=1000
```

```python
false=False
true=True
since = datetime(2019, 8, 15)#date since when the data should be kept
```

```python
survey_response_data = json.load(open('SurveyResponse.json'))
responses = []
for r in survey_response_data:
    if 'received' in r and dateparser.parse(r['received']).date()>= since.date():
        try:
            responses.append({'SurveyID':r['value']['SurveyID'],
                              'UUID': r['uid'],
                              'Response':r['value']['Response']})
        except Exception as e:
            print(r)
```

```python
df = pd.io.json.json_normalize(responses, sep='_')
df.columns = df.columns.map(lambda x: x.split("_")[-1])
df.set_index(['UUID','SurveyID'], inplace=True)
```
